```python
import json, requests
import numpy as np
import pandas as pd

from bs4 import BeautifulSoup
```

```python
base = "https://www.springfieldspringfield.co.uk/tv_show_episode_scripts.php"
res = requests.get(base)
res.status_code
```

```python
soup = BeautifulSoup(res.content, "lxml")
```

```python
print(soup.prettify())
```

```python
page = soup.html
```

```python
script_buttons = page.find_all("a",  {"class":"script-list-item"})
```

```python
script_buttons[0].attrs["href"]
```

```python
page_1_url_list = []
for script in script_buttons:
    script_dict = {}
    script_dict["show"] = script.text
    script_dict["url"] = script.attrs["href"]
    page_1_url_list.append(script_dict)
```

```python
page_1_url_list
```

```python
page_link_span = page.find("div", {"class": "pagination"})
```

```python
page_url_list = page_link_span.find_all("a")
```

```python
page_link_list = []
for page in page_url_list:
    page_dict = {}
    page_dict["page_number"] = page.text
    page_dict["page_url"] = page.attrs["href"]
    page_link_list.append(page_dict)
```
