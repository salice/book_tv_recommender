```python
ia = IMDb()
info_list = ["release dates", "episodes", "reviews", "keywords"]
show_info = []
for show in show_list["show_names"][:2]:
    search_dict = {}
    search = ia.search_movie(show)[0]
    search.get_current_info()
    ia.update(search, info=info_list)
    search_dict["show_name"] = show
    search_dict["release_dates"] = search.get("release dates")
    search_dict["episodes"] = search.get("episodes")
    search_dict["reviews"] = search.get("reviews")
    search_dict["keywords"] = search.get("keywords")
    show_info.append(search_dict)
show_info
```
